Data Mining Assignment 4: Creating a K means Cluster analysis after PCA variable reduction. 

```python
# data processing
import numpy as np
import pandas as pd

#Pretty printing 
import pprint 
# modeling
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans 
from sklearn import cluster
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
# plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_credit = pd.read_csv("https://raw.githubusercontent.com/Alvin-Hoyeong-Kim/ClassWork/main/Python/Data/credit_default_model_data.csv")
df_credit
```

```python
#dropping two columns as direction stated: 
df_credit = df_credit.drop(['default payment next month'],axis =1)
df_credit = df_credit.drop(['group'],axis =1)
df_credit
```

```python
target = 'default payment next month'
```

```python
X_train, X_test = train_test_split(
    df_credit,  test_size=0.66, random_state=0)
X_train
```

```python
X_scaler = StandardScaler()
X_train_std = X_scaler.fit_transform(X_train)
```

```python
k_vals = [4, 5, 6, 7, 8, 9]
p_vals= [10, 15, 20, 25, 30]
```

Consider all permutations listed below for the number of clusters and the number of principal components
(to be used for clustering) and find out which combination generates the best clustering solution. Use
Silhouette scores to assess clustering solutions. For this step, use a random sample of 10,000 records from the
dataset.
• Number of clusters: 4, 5, 6, 7, 8, 9
• Number of principal components: 10, 15, 20, 25, 30
Select the best number of clusters and principal components based on this analysis. (Pick the set of values that
generate the highest Silhouette score. See the screenshot below.)

```python
X_bin = {}
for p in p_vals:
    pca = PCA(p,random_state=314)
    pca_X_train = pca.fit_transform(X_train_std)
    for x in k_vals:
        km = KMeans(n_clusters=x, random_state=42)
        km.fit_predict(pca_X_train)
        score = silhouette_score(pca_X_train, km.labels_, metric='euclidean')
        X_bin[str(p)+" principals and "+str(x)+" clusters"] = score
```

```python
X_bin2 = {}
X_bin2=(sorted(X_bin.items(), key=lambda x:x[1], reverse=True))

print(X_bin2)
```
